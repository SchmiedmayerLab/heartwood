```python
from heartwood.notebook import NotebookSession

session = NotebookSession(session_id="terra-notebook-example")
session.startup_plan()
```

The readiness and capability views contain no credentials or project-file contents. If setup is incomplete, return to the terminal and run `heartwood`.

```python
session.project_readiness(), session.platform_capabilities()
```

To use the browser beside this notebook, run `heartwood --interface web` in a Terra terminal from this project and keep it running. Then execute the next cell to create an authenticated link for the current Terra runtime.

```python
from IPython.display import Markdown
from IPython.display import display as display_markdown

browser_path = session.web_proxy_url()
if browser_path is None:
    display_markdown(Markdown("Heartwood could not verify a browser proxy route in this kernel."))
else:
    display_markdown(Markdown(f"[Open the Heartwood browser]({browser_path})"))
```

Submit a bounded, read-only first task. Any proposed tools appear in `approval_controls` and are not executed until the complete action set is allowed.

```python
view = session.chat(
    "Inspect the files in this project and summarize their purpose. "
    "Do not modify files or access parent directories."
)
view.chat
```

```python
[
    (action.tool_name, action.summary, action.arguments)
    for control in view.approval_controls
    for action in control.actions
]
```

After reviewing every displayed action, resolve the complete pending set with `session.approve(tool_call_id=view.approval_controls[0].target_id)` or reject it with `session.deny(...)`.
